## 导入工具包

```python
import torchaudio
import torch
from ChatTTS import ChatTTS
import soundfile
from IPython.display import Audio
```

## 载入预训练模型

```python
chat = ChatTTS.Chat()
chat.load_models(compile=False) # 设置为Flase获得更快速度，设置为True获得更佳效果
```

## 输入需语音合成的文字

```python
texts = [
    '学人工智能，应该从Python编程开始',
    'Welcome to the world of Artificial Intelligence, where technology really matters',
    '臣妾要告发熹贵妃私通，秽乱后宫，罪不容诛！',
    '那夜的酒，并不足以让我动情',
    '你还有多少惊喜是朕不知道的',
    '人工智能是本世纪最重要的技术突破，但也为人类带来伦理道德、隐私保护、就业结构这些新问题',
    'Generative Artificial Intelligence, is the most important technology [break]in 21 century, but I think human will be more and more lazy',
]
```

## 语音合成模型推理

```python
wavs = chat.infer(texts)
```

## 试听

```python
Audio(wavs[0], rate=24000)
```
