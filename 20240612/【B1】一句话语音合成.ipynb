## 导入工具包

```python
import torchaudio
import torch
from ChatTTS import ChatTTS
import soundfile
from IPython.display import Audio
```

## 载入预训练模型

```python
chat = ChatTTS.Chat()
chat.load_models(compile=False) # 设置为Flase获得更快速度，设置为True获得更佳效果
```

## 输入需语音合成的文字

```python
texts = ['我今天真的Feel Good，跑通了语音合成的开源项目，真的特别感谢子豪兄的代码教程！']
```

## 语音合成模型推理

```python
wavs = chat.infer(texts)
```

## 试听

```python
Audio(wavs[0], rate=24000)
```

## 导出保存为音频文件

```python
torchaudio.save("output_b1.wav", torch.from_numpy(wavs[0]), 24000)
```
